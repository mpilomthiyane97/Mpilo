```python
import openai
from google.colab import userdata

#Set OpenAI API key from Colab's secure storage
client = openai.OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

# üìã Smart Note Summarizer Banner
print("üìã Smart Note Summarizer")

#Upload a text file from local system
from google.colab import files
uploaded = files.upload()

# ‚úÖ Process the uploaded file if available
if uploaded:
    #Get file name and read contents
    file_name = list(uploaded.keys())[0]
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()
        print("\n‚úÖ Original Text:\n", text)

        #Let user choose summary length
        summary_length = input("\nSelect summary length (Short/Medium/Detailed): ").strip().lower()

        #Set token limit based on chosen summary level
        max_tokens = 50 if summary_length == "short" else (150 if summary_length == "medium" else 300)

        print("\n‚è≥ Generating summary...")

        try:
            #Use OpenAI Chat API to summarize the uploaded text
            response = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "user", "content": f"Summarize the following text in {summary_length} detail:\n{text}"}
                ],
                max_tokens=max_tokens,
                temperature=0.7
            )
            #Extract and display the summary
            summary = response.choices[0].message.content.strip()
            print("\n‚úÖ Summary:\n", summary)

        except openai.OpenAIError as e:
            #Handle and display any errors from OpenAI
            print(f"\n‚ùå An error occurred during summarization: {e}")
```
